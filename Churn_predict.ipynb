# Data cleaning 

```python
d=["RowNumber","CustomerId","Surname"]
df=df.drop(d,axis="columns")
```

# convert the alphabetic data into numeric data.

```python
le=LabelEncoder()
df.Geography=le.fit_transform(df["Geography"])
df.Gender=le.fit_transform(df["Gender"])
```

# Using train_test_split allocate the data as you models need and the data availability

```python
x=df.drop("Exited",axis="columns")
y=df.Exited
```

```python
x_train, x_test, y_train, y_test=tts(x,y,test_size=0.2)
```

# find the optimal and efficient model

```python
model=[RandomForestClassifier(),LogisticRegression()]
for i in model:
    model=i.fit(x_train,y_train)
    print(model.score(x_train,y_train))
```

```python
#this shows us the Randomforestclassifier works better so let us predict it with test data set
```

```python
mod=RandomForestClassifier()
mod.fit(x_train,y_train)
y_pre=mod.predict(x_test)
```

```python
mod.score(x_test,y_test)
```
